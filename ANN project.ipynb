```python
import pandas as pd
import librosa 
import numpy as np
metadata=pd.read_csv("UrbanSound8K/metadata/UrbanSound8k.csv")
audio_dataset_path='UrbanSound8K/audio/'
```

```python
metadata.shape[0]#  number of audio in dataset
```

```python
metadata.head()
```

```python
metadata["class"].value_counts()#10 classes
```

```python
 def feature_extractor(filename):
     audio,fs=librosa.load(filename)
     mfcc_feature_matrix=librosa.feature.mfcc(y=audio,sr=fs,n_mfcc=39)
     mfcc_scaled_feature=np.mean(mfcc_feature_matrix.T,axis=0) #[c1,c2,c3....c13]
     return mfcc_scaled_feature
```

```python
from tqdm import tqdm
import os
tqdm(metadata.iterrows())# just checking how it works
```

```python
extracted_feature=[]#[[audio1 features],[audio2 features]....]
for index_num,row in tqdm(metadata.iterrows()):
    file_name=os.path.join(os.path.abspath(audio_dataset_path),'fold'+str(row["fold"])+"/",str(row["slice_file_name"]))
    final_class_labels=row["class"]
    data= feature_extractor(file_name)
    extracted_feature.append([data,final_class_labels])
```

```python
extracted_feature[0]
```

```python
df=pd.DataFrame(extracted_feature,columns=["feature","class"])
```

```python
df.head()
```

```python
#expending features into particular list 
feature_df=pd.DataFrame(df["feature"].tolist())
```

```python
feature_df.head()
```

```python
class_df=pd.DataFrame(df["class"],columns=["class"])
```

```python
df=pd.concat([feature_df,class_df],axis=1)
```

```python
df.info()
```

```python
X=df.iloc[:,:39]
Y=df["class"]
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
le=LabelEncoder()
Y=le.fit_transform(Y)
Y=to_categorical(Y)
Y
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=42)
```

```python
X_train.shape
Y_train.shape
Y_train
```

```python
#ann will be used
import tensorflow
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dropout,Input
from tensorflow.keras.optimizers import RMSprop
```

```python
model=Sequential()
model.add(Input(shape=(39,)))
model.add(Dense(256,activation="relu"))
model.add(Dropout(.25))
model.add(Dense(128,activation="relu"))
model.add(Dropout(.25))
model.add(Dense(64,activation="relu"))
model.add(Dropout(.25))
model.add(Dense(10,activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer=RMSprop(),loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_train,Y_train,verbose=2,epochs=100,batch_size=100,validation_data=(X_test,Y_test))
```

```python
test_accuracy=model.evaluate(X_test,Y_test)
test_accuracy[1]
```

```python
filename="UrbanSound8k/unknown.wav"
prediction_feature=feature_extractor(filename)
prediction_feature=prediction_feature.reshape(1,-1)
pred=model.predict(prediction_feature)
predicted_class = np.argmax(pred, axis=1)
print(predicted_class)
label=le.inverse_transform(predicted_class)
print(label)
```
